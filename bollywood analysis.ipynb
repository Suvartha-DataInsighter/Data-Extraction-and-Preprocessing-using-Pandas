### Which are top 10 movies with maximum return on investment(ROI) calculate return on investment as (BoxOfficeCollection-Budget)/Budget

```python
dframe['ROI']=(dframe['BoxOfficeCollection']-dframe['Budget'])/dframe['Budget']
print(dframe)
dframe.groupby(["MovieName"])['ROI'].max().head(10)
```

### Do the movies have ROI if they get released on festive seasons or long weekened?Calculate the average ROI for different release times

```python
#df_7=dframe[(dframe["ReleaseTime"]=='LW')|dframe["ReleaseTime"=="FS"]
dframe.groupby("ReleaseTime")['ROI'].mean()
```
